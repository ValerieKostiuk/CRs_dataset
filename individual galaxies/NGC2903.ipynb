### measurements reliability

#### Spiral corotation

- Scarano & Lepine (2013) --- *metallicity method* ($153 \pm 25$ arcsec)

> The authors notes that their fitting model (polenomial) found metallicity break both on data sample of Pilyigin (2002) and Zaritsky (1996), making this point strong. Besides, they point out the metalicity break beyond 9 kpc (285 arcsec)

- Marchuk et al.(2024) --- *width method* ($159.2 \pm 16.4$ and $181.6 \pm 9.6$ arcsec)

- McCall (1982) --- *metallicity method* ($139.8$ arcsec)

> The accuracy can not be estimated since in this paper there is no any graphics or notes on the reliability. 

- Abdeen et al. (2020) --- *offset method* ($122 \pm 32$ and $153.75 \pm 20.25$ arcsec)

>The authors traced the spirals on images at different wavelength. They used two different scripts to trace arms. The fisrt one is PYTHON OL, it enables to generate a synthetic logarithmic spiral. ''Since the overlaying synthetic spiral traces are logarithmic,
if the actual spiral structure of the galaxy has intrinsic deviations from being logarithmic, the overlaying would not be accurate.''
The second one is Spirality (spiral arm-count-script), which is also fits logarithmic spiral, also it is capable of measuring the number of arms. These two measurements relate for different spiral tracing methods. Note, the reliability of them depends on whether those spirals are indeed logarithmic.

- Elmegreen & Elmegreen (1995) --- *morphological method* ($153 \pm 16$ arcsec)

> the corotation radius were estimated used the assumption that it locates midway out the disk (0.4-0.6 R25) or near the endpoint of highly symmetric part of the spiral arms. Can be considred as not primary method

- Roberts et al. (1975) --- *morphological method* ($140$ arcsec)

> The radius where corotation occurs, and consequently the speed at which the pattern rotates, are expected to influence the radial extents of at least three easily observable tracers: (1) the prominent spiral structure, (2) the easily visible disk, and (3) the distribution of H n regions in the original rotation curve study. 

#### Bar corotation

- (https://arxiv.org/pdf/2410.13353) M. Ruiz-Garca et al. (2024) --- *bar torque* ($138 \pm 10$ arcsec (6.7 kpc))

  The length of the bar: 68 arcsec (3.3 kpc)
