### measurements reliability

#### Spiral corotation

- Tamburro et al.(2008) --- *offset method* ($171 \pm 28$ arcsec)

> A good consistency between theoretical and observable profile of angular offset. 



- Rand & Wallin (2004) --- *T-W method* 

> In this paper the authors calculated the pattern speed for southern west spiral as a linear regression for three slits. $\Omega_p = 23 \pm 4$ km/s/kpc. They used BIMA SONG data cube, assuming galaxy to have inclination 65 degree.



#### Bar corotation

- Rand & Wallin (2004) --- *T-W method* 

> The authors calculated the pattern speed of the bar as a linear regression for 8 slits. $\Omega_p = 50 \pm 2$ km/s/kpc. They used BIMA SONG data cube, assuming galaxy to have inclination 65 degree.

- Williams et al. (2021) --- *T-W method*

> Rc = $77 \pm 26$ arcsec related to 50.7 km/s/kpc were estimated used MUSE data of gaseous component Halpha. The author marked this mesurement as the first quality.
 
> These a possible mistake in measurement taken based on stellar component of MUSE data. According to table 2, corotation radius of the bar equal to 34 arcsec (1.9 kpc) relates to pattern speed of 29 km/s/kpc. Besides, as authors note, the linear regression has the second quality. Althogh the stellar component is mor relevant for these method, we have to exclude this measurement.
 
> Rc = $85 \pm 9$ arcsec related to 45.8 km/s/kpc were estimated used ALMA data in CO line
 
- (https://arxiv.org/pdf/2410.13353) M. Ruiz-Garca et al. (2024) --- *bar torque* ($94 \pm 3.5$ arcsec (5.2 kpc))

The length of the bar: 66 arcsec (3.6 kpc)
