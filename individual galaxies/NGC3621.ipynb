### measurements reliability

#### spiral corotation

- Williams et al. (2021) --- *T-W method* ($100\pm5$ arcsec)

In this paper the method was primarly directed to measure the pattern speed and corotation resonances of the bar. Since there is no sign of the bar in NGC3621, this measurement is apperanty relates to the spiral pattern, rotating at 36.3 kpc/s/kpc. The authors estimates the quality flags of linear regression as 2, which means ``Clear multiple slopes visible in the hvi/hxi plot, but otherwise the fit would be a quality flag (1)``. Also notes, that the magnitude of the uncernaties is promissing and add quality to the measurement.

- Scarano & Lepine (2012) --- *metallicity method* ($271 \pm 31$ arcsec)

``The radial oxygen abundance of this galaxy was extensively studied by Bresolin, Kennicutt & Ryan-Weber (2012) using different calibrators for its 73 H ii regions. These authors found a significant flattening in the metallicity gradient of this galaxy after R25 using the same procedure of Zahid & Bresolin (2011). Our procedure agrees with their results and two connected lines and two straight lines with a step fit the data and exhibit an evident break near 8.7 kpc.`` Besides, the aithors point out the poor reliability of the measurement presented in Tamburro. They rather suppose the location of corotation at 6.2 kpc (193 arcsec).

- Tamburro et al.(2008) --- *offset method* ($134 \pm 48$ arcsec)
 
 The authors used SINGS and THINGS data sets to verify the angular offsets between the 24 m and the H i emission tracers. According fig.4 the shape of theoretical curve based ob rotation velocity curve does not fit the observed offset, so this measurement is not accurate, and the location of Rc consistent with this figure is much futher from the center (193 arcsec).
