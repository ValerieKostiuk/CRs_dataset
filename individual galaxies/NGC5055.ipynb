### measurements reliability

#### spiral corotation

- Font et al.(2014) (https://arxiv.org/pdf/1310.3415) --- *F-B method* 

> There are seven resonances, two of them the authors relate to OLR ($90$ and $130$ arcsec). 



- Tamburro et al.(2008) --- *offset method* ($260 \pm 120$ arcsec)
 
> The authors used SINGS and THINGS data sets to verify the angular offsets between the 24 m and the H i emission tracers. According fig.4 the shape of theoretical curve based ob rotation velocity curve fit the observed offset, but the measured values spread around the curve. This approach does not allow us to estimate the corotation location accurately. ``We also note that the pattern speed obtained in this paper for the galaxies NGC 5055 and NGC 5194 disagree with previous independent measurements (e.g. Thornley & Mundy 1997; Zimmer et al. 2004). The fit for NGC 5055 presents large error bars due to the large scatter in the azimuthal offsets (see Figure 4), which may explain the differences.`` 
 
 
 
- Roberts et al. (1975) --- *morphological method* ($140$ arcsec)

> The radius where corotation occurs, and consequently the speed at which the pattern rotates, are expected to influence the radial extents of at least three easily observable tracers: (1) the prominent spiral structure, (2) the easily visible disk, and (3) the distribution of H n regions in the original rotation curve study. 
  
- Elmegreen et al. (1992) --- *morphological method* ($87$ arcsec)

> The authors investigated the resonaces locations in galaxies with three arm using mophological features.(**m=3**)
